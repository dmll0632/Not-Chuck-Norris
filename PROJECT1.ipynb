```python
import requests
import pandas as pd
# Define the URL for the World Bank API
BASE_URL = "http://api.worldbank.org/v2/country/{}/indicator/NY.GDP.MKTP.CD"
COUNTRIES = {"EUU": "European Union", "NAC": "North America", "JPN": "Japan"}
# Parameters for the API request
params = {
    "date": "1980:2023",  # Date range
    "format": "json",     # Response format
    "per_page": 1000      # Number of results per page
}
# Function to fetch GDP data for a given country code
def fetch_gdp_data(country_code):
    url = BASE_URL.format(country_code)
    response = requests.get(url, params=params)
    data = response.json()
    if response.status_code == 200 and data:
        return data[1]
    else:
        return []
# Fetch data for each country
gdp_data = {}
for code, name in COUNTRIES.items():
    gdp_data[name] = fetch_gdp_data(code)
# Process and convert the data into a DataFrame
all_data = []
for country, data in gdp_data.items():
    for entry in data:
        if 'value' in entry and entry['value'] is not None:
            all_data.append({
                "Country": country,
                "Year": entry['date'],
                "GDP": entry['value']
            })
df = pd.DataFrame(all_data)
# Convert year to integer for better handling
df['Year'] = df['Year'].astype(int)
# Sort the DataFrame by Year
df = df.sort_values(by="Year")
# Display the DataFrame
print(df)
# Save DataFrame to a CSV file
df.to_csv("gdp_data.csv", index=False)
```

```python
#df_game = pd.DataFrame(data)
# Filter the DataFrame for years up to 2015
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_game = df_game[df_game['Year'] <= 2015]
# Group by 'Year' and 'Genre' and count the occurrences
genre_trends = df_game.groupby(['Year', 'Genre']).size().unstack().fillna(0)
# Identify the top 5 genres based on total counts
top_genres = genre_trends.sum().nlargest(5).index
# Filter to include only the top 5 genres
genre_trends_top5 = genre_trends[top_genres]
# Optionally, convert counts to proportions for each year
genre_proportions_top5 = genre_trends_top5.div(genre_trends_top5.sum(axis=1), axis=0)
# Plotting the trends for the top 5 genres
plt.figure(figsize=(14, 7))
for genre in genre_trends_top5.columns:
    plt.plot(genre_trends_top5.index, genre_trends_top5[genre], label=genre)
plt.title('Trends in Video Game Genres Over Time (up to 2015)')
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.legend(title='Genre')
plt.show()
```

```python
sales_trends = df_game.groupby(['Year', 'Genre'])[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum()
# Create a DataFrame that sums sales across all years for each genre
total_sales_per_genre = sales_trends.groupby('Genre').sum()
# Sum the sales across all regions to determine the top genres globally
top_genres = total_sales_per_genre.sum(axis=1).nlargest(5).index
# Filter to include only the top 5 genres
sales_trends_top5 = sales_trends.loc[(slice(None), top_genres), :]
# Plotting the trends for the top 5 genres in each region
plt.figure(figsize=(18, 9))
for region in ['NA_Sales', 'EU_Sales', 'JP_Sales']:
    plt.subplot(1, 3, {'NA_Sales': 1, 'EU_Sales': 2, 'JP_Sales': 3}[region])
    for genre in top_genres:
        plt.plot(sales_trends_top5.xs(genre, level='Genre').index,
                 sales_trends_top5.xs(genre, level='Genre')[region],
                 label=genre)
    plt.title(f'Trends in {region[:-6]} Sales by Genre (up to 2015)')
    plt.xlabel('Year')
    plt.ylabel('Sales (Millions)')
    plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```
